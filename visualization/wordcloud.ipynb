```python
import wordcloud 
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

with open("D://result.txt","r",encoding='UTF-8') as f:
    text=f.read()
    
mask = np.array(Image.open('D:/OpenSource.png')) # 定义词频背景
wc = wordcloud.WordCloud(
    background_color=(255,255,255),
    mask=mask, # 设置背景图
    max_words=200, # 最多显示词数
    max_font_size=45 # 字体最大值
)
wc.generate(text)
plt.imshow(wc) # 显示词云
plt.axis('off') # 关闭坐标轴
plt.show() # 显示图像
wc.to_file('3.png')
```
