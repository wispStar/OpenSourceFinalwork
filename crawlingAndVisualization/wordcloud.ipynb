```python
import wordcloud 
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

with open("D://result.txt","r",encoding='UTF-8') as f:
    text=f.read()
    
mask = np.array(Image.open('D:/OpenSource.png')) # 
wc = wordcloud.WordCloud(
    background_color=(255,255,255),
    mask=mask, # 
    max_words=200, # 
    max_font_size=45 # 
)
wc.generate(text)
plt.imshow(wc) # 
plt.axis('off') # 
plt.show() # 
wc.to_file('3.png')
```
